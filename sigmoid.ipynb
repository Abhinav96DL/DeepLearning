```
import numpy as np
from matplotlib import pylab
import pylab as plt

def sigmoid(x):
  s=1/(1+np.exp(x))
  return s

def sigmoid_derivative(x):
    s=sigmoid(x)
    ds=s*(1-s)
    return ds

values=plt.linspace(-10,10,100)
plt.plot(values,sigmoid(values),'r')  
plt.plot(values,sigmoid_derivative(values),'r') 
plt.grid()
plt.show()    
```
